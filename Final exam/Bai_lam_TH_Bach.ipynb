```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('white')
```

### Câu 1 - 3:
#### - Tạo DataFrame df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. Các danh sách này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv
#### - Cho biết số cột, số dòng, các dữ liệu head và tail của DataFrame df.
#### - Tạo lại index cho DataFrame df với giá trị bắt đầu từ 0


```python
cn = pd.read_csv('sinh_vien_CN.csv')
to = pd.read_csv('sinh_vien_TO.csv')
vl = pd.read_csv('sinh_vien_vl.csv')
df = pd.concat([cn,to,vl]).reset_index(drop=True)
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

### Câu 4:
#### Tạo DataFrame df_ket_qua từ tập tin ket_qua.csv

```python
df_ket_qua = pd.read_csv('ket_qua.csv')
df_ket_qua.head()
```

### Câu 5:
#### Tạo DataFrame df_dtb bằng cách nhóm df_ket_qua theo masv và tính điểm trung bình cho mỗi sinh viên.

```python
df_dtb = df_ket_qua.groupby('masv').mean().reset_index()
df_dtb.head()
```

### Câu 6:
#### Trộn df và df_dtb theo masv và có kết quả của df như sau:

```python
df = pd.merge(df, df_dtb, on='masv', how='left')
df.head()
```

### Câu 7:
#### In ra các sinh viên của khoa CN và có diem>8

```python
df_CN_8plus = df[(df['makh']=='CN') & (df['diem']>8)]
df_CN_8plus.head()
```

### Câu 8: 
#### In ra các sinh viên của khoa CN và có diem>8. Sắp xếp tăng theo ten và giảm theo diem

```python
df_CN_8plus.sort_values(['ten', 'diem'], ascending=[True, False]).head()
```

### Câu 9:
#### In ra các nam sinh viên không có học bổng.

```python
df[(df['gioitinh'] == True) & (df['hocbong']==0)].head()
```

### Câu 10:
#### In ra tổng học bổng của mỗi khoa.

```python
df.groupby('makh')['hocbong'].sum().reset_index()
```

### Câu 11:
#### Thống kê học bổng của mỗi khoa.

```python
df[['makh','hocbong']].groupby('makh').agg(['sum', 'max', 'min', 'mean']).reset_index()
```

### Câu 12:
#### Thống kê điểm của mỗi khoa.

```python
df[['makh','diem']].groupby('makh').agg(['count', 'max', 'min', 'mean']).reset_index()
```

### Câu 13:
#### Cho biết các sinh viên có điểm lớn nhất, điểm nhỏ nhất.

```python
df[df['diem'] == df['diem'].min()]
```

```python
df[df['diem'] == df['diem'].max()]
```

### Câu 14:
#### Trong DataFrame df tạo thêm cột ketqua với kết quả là giỏi nếu điểm>=8 và kết quả là khá nếu điểm<8.

```python
df['ketqua'] = df['diem'].apply(lambda x: 'giỏi' if x >= 8 else 'khá')
df[['masv', 'ho', 'ten', 'makh', 'diem', 'ketqua']].head()
```

### Câu 15:
#### Thống kê theo mẫu:

```python
df.groupby(['makh', 'ketqua']).count()['masv'].reset_index().rename(columns={'masv': 'count'})
```
