#### Cho dữ liệu Pokemon.csv => đọc dữ liệu
- Vẽ scatterplot: Phân loại với các điểm không chồng chéo, so sánh Attack (trục x) & Defense (trục y)
- Vẽ scatterplot: So sánh Attack (trục x) & Defense (trục y), có sử dụng thêm regression line với lmplot()
- Vẽ boxplot cho toàn bộ dữ liệu
- Vẽ boxplot không có các cột 'Total', 'Generation', 'Legendary'
- Vẽ histogram: biểu diễn tần suất của cột Attack 
- Vẽ barplot: Nhóm dữ liệu theo Type 1, cho biết mỗi type có mấy mẫu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
sb.set()
```

```python
df = pd.read_csv('../dataset/pokemon.csv', index_col=0)
df.head()
```

```python
plt.figure(figsize=(10,8))
sb.swarmplot(data=df, x='Attack', y='Defense', size=7);
```

```python
sb.lmplot(data=df, x='Attack', y='Defense');
```

```python
sb.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df2 = df.drop(['Total', 'Generation', 'Legendary'], axis=1)
sb.boxplot(data=df2)
plt.xticks(rotation=45)
plt.show()
```

```python
sb.distplot(df['Attack'], kde=False);
```

```python
result = df.groupby('Type 1').count()['HP'].reset_index()
result
```

```python
sb.barplot(data=result, x='Type 1', y='HP')
plt.xticks(rotation=45)
plt.show()
```
