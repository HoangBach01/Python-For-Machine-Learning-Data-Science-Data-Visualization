```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
x = ["Excellent","Very Good","Good","Midium"]
y = [5.0,15.0,20.0,5.0]
df1 = pd.DataFrame({'Type': x, 'Count': y})
df1
```

```python
sb.barplot(data=df1, x='Type', y='Count');
```

```python
chars = "A A A A A A A A A A B B B B B B C C C C C C D D D D D D D D D D E E E E E E E F F F F G G G G G G G H H H H"
lst = chars.split()
num = np.ones(len(lst))
df_2 = pd.DataFrame({'chars': lst, 'num': num})
df2 = df_2.groupby('chars').sum().reset_index()
df2
```

```python
sb.barplot(data=df2, x='chars', y='num');
```

```python
sb.swarmplot(data=df_2, x='chars', y='num');
```
