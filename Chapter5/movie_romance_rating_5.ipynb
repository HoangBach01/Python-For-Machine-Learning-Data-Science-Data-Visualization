# movie_romance_rating_5

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('./movies_data/movies.csv', sep=',', index_col=0)
print(type(movies))
print(movies.shape)
movies.head()
```

```python
ratings = pd.read_csv('./movies_data/ratings.csv', sep=',', parse_dates=['timestamp'])
print(type(ratings))
print(ratings.shape)
print(ratings.head())
```

```python
movies_rating = ratings[['movieId', 'rating']]
```

```python
rating = movies_rating.groupby('movieId').mean()
```

```python
rating.head()
```

```python
rating.info()
```

```python
result = pd.concat([movies, rating], axis=1, sort=False)
result.head()
```

```python
result.info()
```

```python
movie_romance_rating_5 = result[(result.genres.str.find("Romance")>=0) & (result.rating == 5.0)]
movie_romance_rating_5
```
