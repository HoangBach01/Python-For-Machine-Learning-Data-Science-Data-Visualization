```python
import pandas as pd
```

```python
# Câu 1: Tạo data frame drink có dữ liệu được lấy ra từ tập tin drinks.csv với index_col là cột đầu tiên của dữ liệu. 
# In type, shape, head, tail
drinks = pd.read_csv('drinks.csv', index_col=0)
print(type(drinks))
print(drinks.shape)
print(drinks.columns)
drinks.head()
```

```python
drinks.tail()
```

```python
# Câu 2: Thống kê trung bình beer_servings nhóm theo 'continent'
drinks.groupby('continent').beer_servings.mean()
```

```python
# Câu 3: Thống kê tổng quát wine_servings nhóm theo 'continent'
drinks.groupby('continent').wine_servings.describe()
```

```python
# Câu 4: Thống kê trung bình cho tất cả các cột theo 'continent'
drinks.groupby('continent').mean()
```

```python
# Câu 5: Thống kê median cho tất cả các cột theo 'continent'
drinks.groupby('continent').median()
```

```python
# Câu 6: Thống kê trung bình, min, max của spirit_servings nhóm theo 'continent' dùng spirit_servings.agg(['mean', 'min', 'max'])
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

```python
# Câu 7: Sắp xếp dữ liệu giam dần theo cột beer_servings. In 5 country có beer_servings lớn nhất, nhỏ nhất
drinks_des = drinks.sort_values(by='beer_servings', ascending=False)
```
