# Chapter 5 - exercise 2: Merge dữ liệu

```python
import pandas as pd
```

## Cung cấp các dictionary sau:

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

## Yêu cầu:
1. Tạo data1, data2, data3 từ 3 dictionary trên. In data1, data2, data3
2. Tạo all_data bằng cách gộp data1 và data2 theo dòng. In all_data
3. Tạo all_data_col bằng cách gộp data1 và data2 theo cột. In all_data_col.
4. Tạo all_data_3 bằng cách gộp all_data và data3, với cột chung là 'subject_id'. In all_data_3
5. In thông tin của all_data_3
6. In thống kê chung của all_data_3

```python
# Câu 1: Tạo data1, data2, data3 từ 3 dictionary trên. In data1, data2, data3
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])
```

```python
data1
```

```python
data2
```

```python
data3
```

```python
# Câu 2: Tạo all_data bằng cách gộp data1 và data2 theo dòng. In all_data
all_data = pd.concat([data1, data2])
all_data
```

```python
# Câu 3: Tạo all_data_col bằng cách gộp data1 và data2 theo cột. In all_data_col.
all_data_col = pd.concat([data1, data2], axis = 1)
all_data_col
```

```python
# Câu 4: Tạo all_data_3 bằng cách gộp all_data và data3, với cột chung là 'subject_id'. In all_data_3.
all_data_3 = pd.merge(all_data, data3, on='subject_id')
all_data_3
```

```python
# câu 5: In thông tin của all_data_3
all_data_3.info()
```

```python
# Câu 6: In thống kê chung của all_data_3
all_data_3.describe()
```

```python
all_data_4 = pd.merge(all_data, data3, on='subject_id', how='left')
all_data_4
```

```python
all_data_4 = pd.merge(all_data, data3, on='subject_id', how='right')
all_data_4
```

```python
all_data_4 = pd.merge(all_data, data3, on='subject_id', how='outer')
all_data_4
```
