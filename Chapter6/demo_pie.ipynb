```python
import matplotlib.pyplot as plt
```

```python
x = [46, 37, 35, 53]
labels  = ["10A1", "10A2", "10A3", "10A4"]
```

```python
plt.pie(x, labels = labels)
plt.show()
```

```python
plt.pie(x, labels = labels, counterclock=False, shadow=True, autopct='%.0f%%')
plt.title("Số học viên trong lớp")
plt.legend(loc=4)
plt.show()
```
