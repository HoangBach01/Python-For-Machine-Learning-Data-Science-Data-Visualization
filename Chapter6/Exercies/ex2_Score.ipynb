```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1], 
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3], 
            'postTestScore': [25, 94, 57, 62, 70]}
```

```python
# 1. Tạo DataFrame df từ dữ liệu trên. In df. 
df = pd.DataFrame(raw_data)
df
```

```python
# 2. Tạo một Scatterplot của preTestScore và postTestScore, với kích thước s của mỗi điểm được xác định theo age. Thiết lập xlabel, ylabel và title. 
plt.style.use('default')
plt.xkcd()
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'])

plt.xlabel('Pre-test Score')
plt.ylabel('Post-test Score')
plt.title('Scatter Plot between Pre & Post test Score')

plt.show()
```

```python
# 3. Tạo một Scatterplot của preTestScore và postTestScore. Lần này kích thước s sẽ gấp 4.5 lần postTestScore. Màu c được xác định theo giới tính (df.female). Thiết lập xlabel, ylabel và title.

plt.xkcd()
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*4.5, c=df['female'])

plt.xlabel('Pre-test Score')
plt.ylabel('Post-test Score')
plt.title('Scatter Plot between Pre & Post test Score')

plt.show()
```
