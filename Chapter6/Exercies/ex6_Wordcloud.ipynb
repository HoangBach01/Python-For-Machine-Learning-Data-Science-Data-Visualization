## Part 1:
- Cho dữ liệu Youtube04-Eminem.csv là dữ liệu được lấy từ UCI Machine Learning Repository, trong đó chứa các YouTube comment cho các video của các nghệ sỹ nổi tiếng.
- Hãy bỏ các STOPWORD
- Vẽ WordClouds cho dữ liệu này

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
data = pd.read_csv('../youtube/Youtube04-Eminem.csv', encoding='latin-1')
text = ""
for i in data['CONTENT']:
    text = text + " " + i
```

```python
wc = WordCloud(background_color='black', max_words=1000, stopwords=stopwords)
wc.generate(text)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Part 2: 
- Sử dụng vietnamese stop word từ : https://github.com/stopwords/vietnamese-stopwords => tạo thành set các stop words
- Đọc file ngon_tu_quang_cao.txt => đưa nội dung vào biến text
- Bổ sung thêm một số từ không quan trọng vào stopwords
- Vẽ wordclouds
- Chọn hỉnh làm wc_mask phù hợp => vẽ wordclouds với wc_mask

```python
stopwords_vn = set()
f = open('vietnamese-stopwords.txt', 'r', encoding='utf-8')
for line in f:
    word = f.readline()
    stopwords_vn.add(word.replace('\n',''))
f.close()
```

```python
df_stopword = pd.read_csv('vietnamese-stopwords.txt', encoding='utf-8', header=None)
stopwords_vn_df = set(map(str,df_stopword[0]))
```

```python
f2 = open('../ngon_tu_quang_cao.txt')
text2 = ""
for line in f2:
    text2 = text2 + line.replace('\n',' ')
f2.close()
```

```python
from PIL import Image
```

```python
wc_mask = np.array(Image.open('../heart.png'))
```

```python
wc = WordCloud(background_color='white', max_words=1000, stopwords=stopwords_vn_df, mask=wc_mask)
wc.generate(text2)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
