```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

line1 = plt.scatter(x, y)#, align='center')

line2 = plt.scatter(x2, y2, color='g')#, align='center')

plt.legend([line1, line2], ['Dot 1', 'Dot 2'], loc = 4)

plt.title('Scatter')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
