```python
%config IPCompleter.greedy = True
import numpy as np
import pandas as pd
```

```python
# Câu 1: Tạo Data Frame từ dictionary như sau:
'''
dic_1 = {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}
In data frame này
'''
dic_1 = {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}
df1 = pd.DataFrame(dic_1)
df1
```

```python
# Câu 2: Tạo Data Frame df2 từ dictionary exam_data, với index của data frame là lables như sau:
'''
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
'''
# In data frame vừa tạo và in info() của data frame
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5,np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df2 = pd.DataFrame(exam_data, index = labels)
df2
```

```python
df2.info()
```

```python
# Câu 3: Tạo data frame mới từ data frame df2, chỉ chứa 2 cột là name và score.  
# In type, shape của data frame này. In head của data frame này
df3 = df2[['name', 'score']]
print(type(df3))
print(df3.shape)
```

```python
df3.head()
```

```python
# Câu 4: Tạo data frame mới từ data frame df2, chỉ chứa 2 cột là name và score, và chỉ có các dòng 1, 3, 5, 6
# In data frame này
df4 = df2.iloc[[1, 3, 5, 6], [1, 3]]
df4
df4 = df2.iloc[[1, 3, 5, 6], [0,2]]
df4
```

```python
df4i = df2.loc[["b", "d", "f", "g"], ["name", "score"]]
df4i
```

```python
# Câu 5: Từ data frame df2, In các dòng có dữ liệu null
df2[df2['score'].isnull()]
```

```python
# Câu 6: Từ data frame df2, In các dòng có score > 15 và <=20
df2[(df2['score'] > 15) & (df2['score'] <= 20)]
```

```python
# Câu 7: Cập nhật điểm ở dòng 'd' thành 18
df2.loc['d', 'score'] = 18
df2.head()
```

```python
# Câu 8: In thống kê chung của df2
df2.describe(include = 'all')
```

```python
# Câu 9: Cho biết diểm mấy có tần suất xuất hiện nhiều nhất? In những dòng này. Cho biết có bao nhiêu dòng thỏa điều kiện?
mark = df2['score'].mode()
print(mark[0])
sc1 = df2[(df2['score']==mark[0])]
sc1
```

```python
print('Số dòng thỏa điều kiện:')
sc1.shape[0]
```

```python
# Câu 10: Thêm dòng k có nội dung như sau: [1, 'Suresh', 'yes', 15.5] vào df2
df2.loc['k'] = ['Suresh', 15.5, 1, 'yes']
df2
```

```python
# Câu 11: Thêm dòng l có nội dung như sau: [2, 'Janny', 'yes', 12.5] vào df2. 
# In tail của df2. Xóa bỏ dòng l của df2. In lại tail.
df_temp = pd.DataFrame({'name': ['Janny'],
        'score': [12.5],
        'attempts': [2],
        'qualify': ['yes']}, columns=["name", "score", "attempts", "qualify"])
df_temp
```

```python
df2.append(df_temp)
df2.tail()
```

```python
df3 = df2.drop(['k'])
df3.tail()
```

```python
# Câu 12: Sắp xếp df2 tăng dần theo điểm
df4 = df3
df4['score'] = df4['score'].apply(pd.to_numeric)
df4 = df4.sort_values(by='score')
df4
```

```python
df4 = df3[['score','attempts']]
print(df4)
print(df4.apply(np.sum,axis=0))
print(df4.apply(np.sum,axis=1))
```

```python
df4 = pd.DataFrame({'point':['1','2','3']})
print(df4)
print(df4.sum())
df4['point'] = df4['point'].apply(pd.to_numeric)
print(df4.sum())
```

```python
# Câu 13: Thêm cột result căn cứ dữ liệu của cột 'score', nếu >=10 thì giá trị = 1, ngược lại = 0
df2['result'] = df2['score'].map(lambda x: 1 if x >= 10 else 0)
df2
```

```python
# Câu 14: Trong df2, thay tên 'Emily' thành 'Samantha'
df2['name'] = df2['name'].replace('Emily', 'Samantha')
df2
```

```python
# Câu 15: Duyệt df2, in name, score, result: nếu = 1 thì in 'Pass', ngược lại thì in 'Fail'
for index, row in df2.iterrows():
    print(row['name'],' - Score:', row['score'], ' - Result:', "Pass" if row['result'] == 1 else "Fail")
```
