### Tạo data frame euro12 tử dữ liệu trên. In euro2: type, shape, danh sách các cột

```python
euro12 = pd.read_csv('euro2012.csv', sep=',')
print(type(euro12))
print(euro12.shape)
print(euro12.columns)
euro12
```

```python
# Câu 1: In thông tin cột Goals
euro12['Goals']
```

```python
# Câu 2: Có bao nhiêu đội tham gia Euro2012?
euro12.shape[0]
```

```python
# Câu 3: In thông tin của Euro2012
euro12.info()
```

```python
# Câu 4: Tạo 1 data frame mới từ euro12 có tên là discipline chỉ chứa 3 cột 'Team', 'Yellow Cards', 'Red Cards'
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
# Câu 5: Sắp xếp discipline giảm dần theo 2 cột 'Red Cards', 'Yellow Cards'
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

```python
# Câu 6: Tính trung bình Yellow Cards
round(discipline['Yellow Cards'].mean())
```

```python
# Câu 6: Lọc các đội đã ghi hơn 6 bàn thắng
euro12[euro12.Goals > 6]
```

### Step 11. Select the teams that start with G

```python
# Câu 7: In các đội mà tên bắt đầu bằng 'G' dùng str.startswith('G')
euro12[euro12.Team.str.startswith('G')]
```

```python
# Câu 8: In 7 cột đầu của euro12
euro12.iloc[: , 0:7]
```

```python
# Câu 9: In tất cả các cột trừ 3 cột cuối
euro12.iloc[: , :-3]
```

```python
# Câu 10: In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']]
```

```python
# Câu 11: In các cột Chỉ hiễn thị 'Team','Shooting Accuracy' từ 'England', 'Italy', 'Russia'
euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team','Shooting Accuracy']]
```
